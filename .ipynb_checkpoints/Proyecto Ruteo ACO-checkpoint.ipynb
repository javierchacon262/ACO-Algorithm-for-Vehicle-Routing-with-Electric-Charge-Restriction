```python
#Funcion que extrae las coordenadas del vector de nodos veCoord
def extractCoord(veCoord):
    veCoordExp = []
    veCoordExp.clear()
    l = 0
    for i in veCoord:
        #print("\n Nueva: ")
        #print(i[0])
        #print("\n")
        veCoordExp.append({})
        k = 0
        for j in i:
            #print("l: " + str(l) + "\n")
            #print("k: " + str(k) + "\n")
            veCoordExp[l].update({str(k) : [float(j[1]), float(j[2])]})
            k += 1
        l += 1
    return veCoordExp
```

```python
#Funcion que extrae demandas, ventanas de tiempo minima y maxima, y tiempos de servicio
def extractDemand(veCoord):
    veDemand = []
    veDemand.clear()
    l = 0
    for i in veCoord:
        veDemand.append({})
        k = 0
        for j in i:
            veDemand[l].update({str(k) : [float(j[3]), float(j[4]), float(j[5]), float(j[6])]})
            k += 1
        l += 1
    return veDemand
```

```python
def euclidean(a, b):
    return math.sqrt(pow(a[1] - b[1], 2) + pow(a[0] - b[0], 2))
```

```python
#Funcion de buscar maximo en lista de diccionarios
def findMaxDict(dic, vecRest):
    vecMaxRest = []
    vecMaxRest.clear()
    for i in vecRest:
        if type(i) != float:
            vecMaxRest.append(dic[i][2])
    return max(vecMaxRest)
```

```python
#Funcion que evalua la necesidad de insertar en la ruta una estacion de carga electrica
def lookCharge(coord, estaciones):
    veDist = []
    for i in range(len(estaciones)):
        if coord != estaciones[str(i)]:
            veDist.append(euclidean(coord, estaciones[str(i)]))
            #print(str(veDist[-1]), "\t\t", str(estaciones[i]), "\n")
    menor = min(veDist)
    minInd = veDist.index(menor)
    return [minInd, estaciones[str(minInd)], menor]
```

```python
#Funcion para tomar decision en funcion de la carga restante
def mercaRest(carro, nodoSig, demandas):
    if carro.cargaM >= demandas[str(nodoSig)][0]:
        flagCarga = True
    else:
        flagCarga = False
    return flagCarga
```

```python
#Funcion de Evaluar decisiones sobre ventanas de tiempo
def EvalTiempo(carro, nodo2, demandas, vecSol, dist1, v):
    flagOnTime = False
    flagSup = False
    flagDep = False
    #SUMA DE TIEMPO PARA CUADRAR CON LAS VENTANAS DE TIEMPO
    #Segunda Decicion A: Ventanas de Tiempo
    if (demandas[str(nodo2)][1] - (dist1 / v) - carro.time) < 0:
        #Verifica que el tiempo que le toma en ir de un cliente a otro este dentro de la ventana de tiempo de ese nodo
        carro.time += (dist1 / v)
    else:
        #cuando la condicion es mayor a cero significa que el carro debe esperar por el cliente para que abra,
        #El carro espera en el nodo anterior
        carro.time += (dist1 / v) + (demandas[str(nodo2)][1] - (dist1 / v) - carro.time)
    
    if carro.time >= demandas[str(nodo2)][1] and carro.time <= demandas[str(nodo2)][2]:
        #cosas
        flagOnTime = True
    elif carro.time > demandas[str(nodo2)][2]:
        #Cuando esta por encima de la ventana de tiempo
        #Pregunta si el tiempo del carro es mayor a la ventana de tiempo mayor
        #Manda de Ultimos los nodos a los que no alcanza a llegar pero cuyas ventanas superiores de tiempo
        #son menores a la del nodo con ventana de tiempo maxima
        if carro.time < findMaxDict(demandas, vecSol):    
            #cosas ventana de tiempo superior
            flagSup = True
        else:
            #Si el tiempo acumulado del carro supera todas las ventanas de tiempo, se viaja al deposito
            #y sale un nuevo carro
            flagDep = True
            #Se aumenta el contador de carros y se crea el siguiente, 
            #Super importante actualizar los nodos y los coords
    return [flagOnTime, flagSup, flagDep]
```

```python
#Funcion para tomar decisiones sobre la bateria
def energiaRest(carro, nodo1, nodo2, dist1, nodes, estNodo1, estNodo2, conRate, recRate, cargaMax, v, costoVeh, costoDis):
    ############################################################################################################   
    #Coordenadas de los nodos
    coord1 = nodes[str(nodo1)]
    coord2 = nodes[str(nodo2)]
    #Energia actual del carro
    eA = float(carro.cargaE)
    #Energia hasta el siguiente nodo
    eNN = dist1 * conRate
    #Energia hasta la estacion mas cercana del siguiente nodo
    eNE = estNodo2[2] * conRate
    #Energia hasta la estacion mas cercana del nodo actual
    eAE = estNodo1[2] * conRate
    #Coordenadas del deposito
    depCoord = nodes['0']
    #distancia estacion nodo actual al deposito
    print("energia: linea 19: ", estNodo1[1])
    distEstDep = (costoVeh) + (costoDis * euclidean(estNodo1[1], depCoord))
    #Energia de la estacion nodo actual al deposito
    eAEDEP = distEstDep * conRate
    #tiempo
    time = 0
    ### flags ###
    flagEner = False
    flagEst = False
    flagDep = False
    #Tomas de Decision
    print("eA: ", type(eA))
    print("eNN: ", type(eNN))
    print("eNE: ", type(eNE))
    if eA - eNN - eNE > 0:
        flagEner = True
        eA -= eNN
    elif eA - eAE > 0:
        eA -= eAE
        distEAENN = euclidean(estNodo1[1], coord2)
        eNN = distEAENN * conRate
        time += distEAENN / v
        flagEst = True
        if eA - eNN - eNE < 0:
            #cargar lo necesario
            dif = eNN + eNE - eA
            if dif <= cargaMax - eA:
                eA += dif - eNN
                time += dif/recRate
                flagEner = True
                #Luego viaja normal
                #cosas de las ventanas de tiempo
            elif eA - eAEDEP < 0:
                dif = eAEDEP - eA
                if dif <= cargaMax - eA:
                    eA += dif - eAEDEP
                    time += dif/recRate
                    flagDep = True
                    nodo2 = 0
                    #Luego viaja al deposito
                else:
                    print("Error de ruta 1, Auto Estancado")
                    raise KeyboardInterrupt
            elif eA - eAEDEP > 0:
                eA -= eAEDEP
                flagDep = True
                nodo2 = 0
                #viaja al deposito
        else:
            eA -= eNN
            flagEner = True
    else:
        #print("Error de ruta 2, Auto Estancado")
        raise KeyboardInterrupt
    return [flagEner, flagEst, flagDep, nodo2, eA, time]
```

```python
#Evaluacion de las restricciones iniciales del problema
def SolInicial(numNodos, cargaM, cargaE, conRate, recRate, velocity, nomInst, nodes, estaciones, demandas, costoVeh, costoDis):
    
    #Inicializar vector de nodos de donde escoger
    vecSol = []
    vecSol.clear()
    for i in range(1, numNodos+1):
        vecSol.append(str(i))
    #Inicializar vector de nodos visitados
    vecVisited = []
    v = float(velocity)
    #Vector de carros que crecera dinamicamente, dependiendo de la instancia
    veCars = []
    #contgador de carros utilizados
    conCars = 0
    #Carro inicial al vector de carros
    veCars.append(Vehiculo(conCars, cargaE, cargaM))
    print("\n carga de mercancia inicial del carro: " + str(conCars) + "  =  " + str(veCars[conCars].cargaM) + "\n")
    i = 0
    nodo1 = None
    nodo2 = None
    dist = 0
    time = 0
    cargaE = float(cargaE)
    recRate = float(recRate)
    conRate = float(conRate)
    cargaM = float(cargaM)
    flagDep = True
    while i < numNodos:
        print("\nNodo: ", i, "\n")
        #Decide si empieza desde el deposito en caso de haber viajado alla o del ultimo nodo visitado
        if flagDep:
            #Deposito como nodo inicial
            nodo1 = 0
            coord1 = nodes[str(nodo1)]
            vecVisited.append(nodo1)
            flagDep = False
        else:
            nodo1 = vecVisited[-1]
            coord1 = nodes[str(nodo1)]
        if len(vecSol) > 0:
            nodo2 = int(random.choice(vecSol))
        else:
            flagDep = True
        #revisa que el nodo2 seleccionado no este entre los visitados
        flagIsIn = True
        while flagIsIn:
            if nodo2 in vecVisited and len(vecSol) > 0:
                nodo2 = int(random.choice(vecSol))
                flagIsIn = True
            else:
                flagIsIn = False
        #Solo averigua las coordenadas de nodo2 sabiendo que es algun nodo
        if nodo2 != None:
            coord2 = nodes[str(nodo2)]
            flagCarga = mercaRest(veCars[conCars], nodo2, demandas)
        else:
            flagCarga = False
            
        #Primera Decicion: Carga de Mercancia
        if flagDep == False and flagCarga:
            flagDep = False
            print("\n nodo1: " + str(nodo1) + "\n")
            #print("\n coordenadas nodo 1: ", coord1, "\n")
            print("\n nodo2: " + str(nodo2) + "\n")
            #print("\n coordenadas nodo 2: ", coord2, "\n")
            
            #FUNCION OBJETIVO:
            ####################################################################################################
            dist1 = (costoVeh) + (costoDis * euclidean(coord1, coord2))
            ####################################################################################################   
            #Buscar Estacion de carga mas cercana a nodo1
            estNodo1 = lookCharge(coord1, estaciones)
            print("Linea 68: estNodo1: ", estNodo1)
            #Buscar Estacion de carga mas cercana a nodo2
            estNodo2 = lookCharge(coord2, estaciones)
            print("Linea 68: estNodo2: ", estNodo2)
            ####################################################################################################
            energiaR = energiaRest(veCars[conCars], nodo1, nodo2, dist1, nodes, estNodo1, estNodo2, conRate, recRate, cargaE, v, costoVeh, costoDis)
            ####################################################################################################
            #Segunda Decicion B: Ventanas de Tiempo
            #Cuando esta dentro de la ventana de tiempo
            if energiaR[0] and not energiaR[1] and not energiaR[2]:
                
                veCars[conCars].eA = energiaR[4]
                #Deciciones sobre ventanas de tiempo
                flagsTime = EvalTiempo(veCars[conCars], nodo2, demandas, vecSol, dist1, v)
                if flagsTime[0] and not flagsTime[1] and not flagsTime[2]:
                    #viaja normal al siguiente nodo
                    flagDep = False
                    dist += dist1
                    veCars[conCars].cargaM -= demandas[str(nodo2)][0]
                    veCars[conCars].time += demandas[str(nodo2)][0]
                    vecVisited.append(nodo2)
                    n = vecSol.pop(vecSol.index(str(nodo2)))
                    i += 1
                    
                elif not flagsTime[0] and flagsTime[1] and not flagsTime[2]:
                    #Evalua otro nodo, salir de la decision a la siguiente iteracion
                    print("###########################################################################################")
                    print("\n Lista de Nodos Nueva por ventana de tiempo SUPERIOR: " + str(findMaxDict(demandas, vecSol)) + "\n\n")
                    print("###########################################################################################")
                    veCars[conCars].time -= dist1 / v
                    nodoUlt = vecSol.pop(vecSol.index(str(nodo2)))
                    vecSol.insert(-1, nodoUlt)
                    
                elif not flagsTime[0] and not flagsTime[1] and flagsTime[2]:
                    #Evaluar energia para viaje al deposito
                    flagDep = True
                    coordDep = nodes['0']
                    distDep = (costoVeh) + (costoDis * euclidean(coord1, coordDep))
                    distNE = (costoVeh) + (costoDis * euclidean(coord1, estNodo1[1]))
                    enerDep = conRate * distDep
                    enerEst = conRate * distNE
                    if veCars[conCars].cargaE > enerDep:
                        dist += distDep
                        veCars[conCars].time += distDep/v
                        veCars[conCars].cargaE -= distDep * conRate
                        conCars += 1
                        veCars.append(Vehiculo(conCars, 79.69, 200))
                    elif veCars[conCars].cargaE > enerEst:
                        dist += distNE
                        veCars[conCars].time += distNE/v
                        veCars[conCars].cargaE -= enerEst
                        vecVisited.append('e' + str(estNodo1[0]))
                        depCoord = nodes['0']
                        distEstDep = (costoVeh) + (costoDis * euclidean(estNodo1, depCoord))
                        enerDep = conRate * distEstDep
                        if veCars[conCars].cargaE < enerDep:
                            dif = enerDep - veCars[conCars].cargaE
                            if dif <= cargaE - veCars[conCars].cargaE:
                                veCars[conCars].cargaE += dif - enerDep
                                veCars[conCars].time += dif/recRate
                                dist += distEstDep
                                veCars[conCars].time += distEstDep / v
                                conCars += 1
                                veCars.append(Vehiculo(conCars, 79.69, 200))
                            else:
                                print("Error de ruta 3, auto estancado")
                        elif veCars[conCars].cargaE > enerDep:
                            dist += distEstDep
                            veCars[conCars].time += distEstDep / v
                            veCars[conCars].cargaE -= distEstDep * conRate
                            conCars += 1
                            veCars.append(Vehiculo(conCars, 79.69, 200))
                    
                    
            elif energiaR[0] and energiaR[1] and not energiaR[2]:
                
                veCars[conCars].eA = energiaR[4]
                veCars[conCars].time += energiaR[5]
                #Decisiones sobre ventanas de tiempo
                flagsTime = EvalTiempo(veCars[conCars], nodo2, demandas, vecSol, dist1, v)
                if flagsTime[0] and not flagsTime[1] and not flagsTime[2]:
                    #viaja normal al siguiente nodo
                    flagDep = False
                    dist += dist1
                    veCars[conCars].cargaM -= demandas[str(nodo2)][0]
                    veCars[conCars].time += demandas[str(nodo2)][0]
                    vecVisited.append('e' + str(estNodo1[0]))
                    vecVisited.append(nodo2)
                    n = vecSol.pop(vecSol.index(str(nodo2)))
                    i += 1
                elif not flagsTime[0] and flagsTime[1] and not flagsTime[2]:
                    #Evalua otro nodo, salir de la decision a la siguiente iteracion
                    print("###########################################################################################")
                    print("\n Lista de Nodos Nueva por ventana de tiempo SUPERIOR: " + str(findMaxDict(demandas, vecSol)) + "\n\n")
                    print("###########################################################################################")
                    veCars[conCars].time -= dist1 / v
                    nodoUlt = vecSol.pop(vecSol.index(str(nodo2)))
                    vecSol.insert(-1, nodoUlt)
                elif not flagsTime[0] and not flagsTime[1] and flagsTime[2]:
                    #Evaluar energia para viaje al deposito
                    flagDep = True
                    coordDep = nodes['0']
                    distDep = (costoVeh) + (costoDis * euclidean(coord1, coordDep))
                    distNE = (costoVeh) + (costoDis * euclidean(coord1, estNodo1[1]))
                    enerDep = conRate * distDep
                    enerEst = conRate * distNE
                    if float(veCars[conCars].cargaE) > enerDep:
                        dist += distDep
                        veCars[conCars].time += distDep/v
                        veCars[conCars].cargaE -= distDep * conRate
                        conCars += 1
                        veCars.append(Vehiculo(conCars, 79.69, 200))
                    elif float(veCars[conCars].cargaE) > enerEst:
                        dist += distNE
                        veCars[conCars].time += distNE/v
                        veCars[conCars].cargaE -= enerEst
                        vecVisited.append('e' + str(estNodo1[0]))
                        depCoord = nodes['0']
                        distEstDep = (costoVeh) + (costoDis * euclidean(estNodo1, depCoord))
                        enerDep = conRate * distEstDep
                        if veCars[conCars].cargaE < enerDep:
                            dif = enerDep - veCars[conCars].cargaE
                            if dif <= cargaE - veCars[conCars].cargaE:
                                veCars[conCars].cargaE += dif - enerDep
                                veCars[conCars].time += dif/recRate
                                dist += distEstDep
                                veCars[conCars].time += distEstDep / v
                                conCars += 1
                                veCars.append(Vehiculo(conCars, 79.69, 200))
                            else:
                                print("Error de ruta 3, auto estancado")
                        elif veCars[conCars].cargaE > enerDep:
                            dist += distEstDep
                            veCars[conCars].time += distEstDep / v
                            veCars[conCars].cargaE -= distEstDep * conRate
                            conCars += 1
                            veCars.append(Vehiculo(conCars, 79.69, 200))
                            
                            
            elif not energiaR[0] and energiaR[1] and energiaR[2]:
                
                veCars[conCars].eA = energiaR[4]
                veCars[conCars].time += energiaR[5]
                flagDep = True
                dist += (costoVeh) + (costoDis * euclidean(coord1, nodes['0']))
                conCars += 1
                veCars.append(Vehiculo(conCars, 79.69, 200))
            
        #Else de la Primera Decicion
        else:
            flagDep = True
            coordDep = nodes['0']
            distDep = (costoVeh) + (costoDis * euclidean(coord1, coordDep))
            distNE = (costoVeh) + (costoDis * euclidean(coord1, estNodo1[1]))
            enerDep = conRate * distDep
            enerEst = conRate * distNE
            if veCars[conCars].cargaE > enerDep:
                dist += distDep
                veCars[conCars].time += distDep/v
                veCars[conCars].cargaE -= distDep * conRate
                conCars += 1
                veCars.append(Vehiculo(conCars, 79.69, 200))
            elif veCars[conCars].cargaE > enerEst:
                dist += distNE
                veCars[conCars].time += distNE/v
                veCars[conCars].cargaE -= enerEst
                vecVisited.append('e' + str(estNodo1[0]))
                depCoord = nodes['0']
                distEstDep = (costoVeh) + (costoDis * euclidean(estNodo1, depCoord))
                enerDep = conRate * distEstDep
                if veCars[conCars].cargaE < enerDep:
                    dif = enerDep - veCars[conCars].cargaE
                    if dif <= cargaE - veCars[conCars].cargaE:
                        veCars[conCars].cargaE += dif - enerDep
                        veCars[conCars].time += dif/recRate
                        dist += distEstDep
                        veCars[conCars].time += distEstDep / v
                        conCars += 1
                        veCars.append(Vehiculo(conCars, 79.69, 200))
                    else:
                        print("Error de ruta 3, auto estancado")
                        raise KeyboardInterrupt
                elif veCars[conCars].cargaE > enerDep:
                    dist += distEstDep
                    veCars[conCars].time += distEstDep / v
                    veCars[conCars].cargaE -= distEstDep * conRate
                    conCars += 1
                    veCars.append(Vehiculo(conCars, 79.69, 200))
    print("\n\n distancia total: " + str(dist))
    print("\n\n contador de carros usados: " + str(conCars+1))
```

```python
class node:
    
    def __init__(self, coord, tipo, demanda, ventana, tServicio):
        self.coord = coord
        self.flagV = False
        self.tipo = tipo
        if self.tipo == 'c':
            self.demanda = demanda
            self.ventana = ventana
            self.tServicio = tServicio
        elif self.tipo == 'e':
            self.demanda = 0
            self.ventana = [0, 999999999]
            self.tServicio = 0
        else:
            print("Error, el tipo de nodo solo puede ser c o e")
            raise KeyboardInterrupt
        
    def getCoord(self):
        return self.coord
    
    def getFlag(self):
        return self.flagV
    
    def setFlag(self, boolVal):
        self.flagV = boolVal
        
    def toString(self):
        return [self.coord, self.flagV]
```

```python
#Clases para el algoritmo de colonia de hormigas
class World:
    
    uid = 0

    def __init__(self, nodes, lfunc, cargaE, conRate, recRate, velocity, estaciones, demandas, costoVeh, costoDis, **kwargs):
        self.uid = self.__class__.uid
        self.__class__.uid += 1
        self.name = kwargs.get('name', 'world{}'.format(self.uid))
        self.description = kwargs.get('description', None)
        #aqui llamo la funcion para seleccionar los nodos factibles 
        self.nodes = nodes
        self.lfunc = lfunc
        self.estaciones = estaciones
        self.edges = self.create_edges()
        self.cargaE = cargaE
        self.recRate = recRate
        self.conRate = conRate
        self.velocity = velocity
        self.demandas = demandas
        self.costoVeh = costoVeh
        self.costoDis = costoDis
        
    #def nodes(self):
    #    return list(range(len(self._nodes)))
    
    def create_edges(self):
        
        edges = {}
        nod = self.nodes + self.estaciones
        for m in nod:
            for n in nod:
                a, b = m.getCoord(), n.getCoord()
                if a != b:
                    edge = Edge(a, b, length=self.lfunc(a, b))
                    edges[m, n] = edge
        return edges
        
    def reset_pheromone(self, level=0.01):
        
        for edge in self.edges.values():
            edge.pheromone = level
        
    def data(self, idx, idy=None):
        try:
            if idy is None:
                return self._nodes[idx].getCoord()
            else:
                return self.edges[idx.getCoord(), idy.getCoord()]
        except IndexError:
            return None    


class Edge:
    
    def __init__(self, start, end, length=None, pheromone=None):
        self.start = start
        self.end = end
        self.length = 1 if length is None else length
        self.pheromone = 0.1 if pheromone is None else pheromone

    #def __eq__(self, other):
    #    if isinstance(other, self.__class__):
    #        return self.__dict__ == other.__dict__
    #    return False
```

```python
class Ant:
    
    uid = 0

    def __init__(self, alpha, beta, merca, energia, costoVeh, costoDis, inst):
        
        self.uid = self.__class__.uid
        self.__class__.uid += 1
        self.world = None
        self.alpha = alpha
        self.beta = beta
        self.start = None
        self.node2 = None
        self.node3 = None
        self.distance = 0
        self.visited = []
        self.unvisited = []
        self.selected = []
        self.traveled = []
        self.merca = merca
        self.mercaMax = cop.deepcopy(merca)
        self.energia = energia
        self.costoVeh = costoVeh
        self.costoDis = costoDis
        self.time = 0
        self.cont = 0

    def initialize(self, world, start=None):
        
        self.world = world
        for i in self.world.nodes:
            i.flagV = False
        if start is None:
            self.start = random.randrange(len(self.world.nodes))
        else:
            self.start = start
        self.distance = 0
        self.visited = [self.start]
        #self.start.flagV = True
        self.unvisited = [n for n in self.world.nodes if n != self.start]
        self.traveled = []
        return self

    def clone(self):
        
        ant = Ant(self.alpha, self.beta)
        ant.world = self.world
        ant.start = self.start
        ant.visited = self.visited[:]
        ant.unvisited = self.unvisited[:]
        ant.traveled = self.traveled[:]
        ant.distance = self.distance
        return ant

    def node(self):
        try:
            return self.visited[-1]
        except IndexError:
            return None

    def tour(self):
        return [self.world.data(i) for i in self.visited]

    def path(self):
        return [edge for edge in self.traveled]
        
    #def __eq__(self, other):
    #    return self.distance == other.distance

    def __lt__(self, other):
        return self.distance < other.distance
    
    def Selec14(self, node1):
        con = 0
        vecN = []
        vecN.clear()
        vecC = []
        vecC.clear()
        if type(node1) == list:
            coord1 = node1[0].getCoord()
        else:
            coord1 = node1.getCoord()
        for i in self.world.nodes:
            ##print(i.flagV)
            if i != self.node() and not i.flagV:
                coord2 = i.getCoord()
                dist = (self.world.costoVeh) + (self.world.costoDis * euclidean(coord1, coord2))
                vecN.append([i, dist])
        long = len(vecN)
        #if long > 0 and long < 1:
        #    long = int(long) + 1
        #elif long > 1:
        #    long = int(long)
        if long > 0:
            for i in range(long):
                vecC.append(vecN.pop(vecN.index(min(vecN, key = lambda x : x[-1]))))
            #print("\nNodos Selec14: ", len(vecC))
            return vecC
        else:
            return []
        
    #Funcion de Evaluar decisiones sobre ventanas de tiempo
    def evalTiempo(self, nodo2, dist2, nodo3, nodes, timeAnt):
        v = self.world.velocity
        flagOnTime = False
        flagSup = False
        flagDep = False
        flagN3 = False
        time = 0
        if type(nodo2) == list:
            vent20 = nodo2[0].ventana[0]
            vent21 = nodo2[0].ventana[1]
            tServ2 = nodo2[0].tServicio
        else:
            vent20 = nodo2.ventana[0]
            vent21 = nodo2.ventana[1]
            tServ2 = nodo2.tServicio
        if nodo3 is not None:
            flagN3 = True
            if type(nodo3) == list:
                vent30 = nodo3[0].ventana[0]
                vent31 = nodo3[0].ventana[1]
                tServ3 = nodo3[0].tServicio
                if type(nodo2) == list:
                    dist3 = (self.world.costoVeh) + (self.world.costoDis * euclidean(nodo2[0].getCoord(), nodo3[0].getCoord()))
                else:
                    dist3 = (self.world.costoVeh) + (self.world.costoDis * euclidean(nodo2.getCoord(), nodo3[0].getCoord()))
            else:
                vent30 = nodo3.ventana[0]
                vent31 = nodo3.ventana[1]
                tServ3 = nodo3.tServicio
                if type(nodo2) == list:
                    dist3 = (self.world.costoVeh) + (self.world.costoDis * euclidean(nodo2[0].getCoord(), nodo3.getCoord()))
                else:
                    dist3 = (self.world.costoVeh) + (self.world.costoDis * euclidean(nodo2.getCoord(), nodo3.getCoord()))
                
        if time + (dist2/v) < vent20:
            #ESPERAR Y VIAJAR
            time += vent20 - (dist2/v) - time
            flagOnTime = True
            #TIEMPO DE SERVICIO
            time += tServ2
            if flagN3:
                if time + (dist3/v) < vent30:
                    #ESPERAR Y VIAJAR NODO 3
                    time += vent30 - (dist3/v) - time
                    flagN3 = True
                    #TIEMPO DE SERVICIO NODO 3
                    time += tServ3
                elif time + (dist3/v) > vent30 and time + (dist3/v) < vent31:
                    #VIAJE Y TIEMPO DE SERVICIO NODO3
                    time += (dist3/v) + tServ3
                    flagN3 = True
                elif time + (dist3/v) > vent31:
                    #DESCARTAR NODO3
                    flagN3 = False
        elif time + (dist2/v) > vent20 and time + (dist2/v) < vent21:
            #VIAJE Y TIEMPO DE SERVICIO
            time += (dist2/v) + tServ2
            flagOnTime = True
            if flagN3:
                if time + (dist3/v) < vent30:
                    #ESPERAR Y VIAJAR NODO 3
                    time += vent3 - (dist3/v) - time
                    flagN3 = True
                    #TIEMPO DE SERVICIO NODO 3
                    time += tServ3
                elif time + (dist3/v) > vent30 and time + (dist3/v) < vent31:
                    #VIAJE Y TIEMPO DE SERVICIO NODO3
                    time += (dist3/v) + tServ3
                    flagN3 = True
                elif time + (dist3/v) > vent31:
                    #DESCARTAR NODO3
                    flagN3 = False
        elif time + (dist2/v) > vent21:
            #DESCARTAR NODO 2 Y NODO 3
            flagSup = True
            flagOnTime = False
            flagN3 = False
            #print("\nlista de nodos:", nodes)
            timeMax = self.nodeMaxVent(nodes).ventana[1]
            if time > timeMax:
                flagDep = True
            
        return([flagOnTime, flagN3, flagSup, flagDep, time])
    
    def nodeMaxVent(self, nodes):
        maxi = 0
        maxN = None
        for i in nodes:
            if type(i) == list:
                if i[0].ventana[1] > maxi:
                    maxi = i[0].ventana[1]
                    maxN = i[0]
            else:
                if i.ventana[1] > maxi:
                    maxi = i.ventana[1]
                    maxN = i
        return maxN
    
    def recargar(self, eA, distNAEst1, distN2, distN2Est2, distN3, distN3Est3):
        cRate = self.world.conRate
        if eA < cRate * (distN2 + distN2Est2):
            dif = self.world.cargaE
            if cRate * (distN2 + distN2Est2) < dif:
                if cRate * (distN2 + distN3 + distN3Est3) < dif:
                    eA = (cRate * (distN2 + distN3 + distN3Est3)) + 1
                else:
                    eA = (cRate * (distN2 + distN2Est2)) + 1
            elif cRate * distNAEst1 < dif:
                eA = (cRate * distNAEst1) + 1
                
        return eA
                
    def energiaRst(self, node2):
        ############################################################################################################   
        eA = cop.deepcopy(self.energia)
        #Cuestiones Nodo Actual
        node1 = cop.deepcopy(self.node())
        coord1 = node1.getCoord()
        #Estacion Nodo Actual
        est1 = self.world.estaciones[self.look_charge(coord1)[0]]
        coordEst1 = est1.getCoord()
        distNAEst1 = (self.world.costoVeh) + (self.world.costoDis * euclidean(coord1, coordEst1))
        #print("\nDistancia nodo actual a estacion mas cercana: ", distNAEst1)
        #Cuestiones Nodo Sig
        if type(node2) == list:
            coord2 = node2[0].getCoord()
        else:
            coord2 = node2.getCoord()
        distN2 = (self.world.costoVeh) + (self.world.costoDis * euclidean(coord1, coord2))
        #print("\nDistancia nodo actual a node2: ", distN2)
        est2 = self.world.estaciones[self.look_charge(coord2)[0]]
        coordEst2 = est2.getCoord()
        distEst1N2 = (self.world.costoVeh) + (self.world.costoDis * euclidean(coordEst1, coord2))
        #print("\nDistancia estacion nodo1 a node2: ", distEst1N2)
        distN2Est2 = (self.world.costoVeh) + (self.world.costoDis * euclidean(coord2, coordEst2))
        #print("\nDistancia node2 actual a estacion2: ", distN2Est2)
        #Energias
        cRate = self.world.conRate
        eNodo2 = (distN2 * cRate) + (distN2Est2 * cRate)
        eEst1 = distNAEst1 * cRate

        vec = self.Selec14(node2)
        if len(vec) > 0:
            if vec[0] == node2:
                if len(vec) > 1:
                    node3 = vec[1]
                    #print("\nnode3: ", node3, "\n")
                else:
                    node3 = None
            else:
                node3 = vec[0]
            if node3 != None:
                if type(node3) == list:
                    coord3 = node3[0].getCoord()
                else:
                    coord3 = node3.getCoord()
                distN3 = (self.world.costoVeh) + (self.world.costoDis * euclidean(coord2, coord3))
                est3 = self.world.estaciones[self.look_charge(coord3)[0]]
                coordEst3 = est3.getCoord()
                distN3Est3 = (self.world.costoVeh) + (self.world.costoDis * euclidean(coord3, coordEst3))
                eNE3 = (distN3 * cRate) + (distN3Est3 * cRate)
        else:
            node3 = None
        #tiempo
        time = 0
        ### flags ###
        flagEner = False
        flagN3 = False
        flagEst = False
        #Tomas de Decision
        if node1.tipo == 'e':
            if distN3 != None:
                eA = self.recargar(eA, distNAEst1, distN2, distN2Est2, distN3, distN3Est3)
            else:
                eA = self.recargar(eA, distNAEst1, distN2, distN2Est2, None, None)
        #print("\nenergia actual verificacion: ", eA)
        #print("\nenergia nodo2: ", eNodo2)
        if eA > eNodo2:
            flagEner = True
            #ver uno mas adelante
            if node3 is not None:
                if eA > (distN2 * cRate) + eNE3:
                    flagN3 = True
                else:
                    flagN3 = False
            else:
                flagN3 = False
            #ver necesidad de cargar
        elif eA > eEst1:
            #print("\nPuede ir a una estacion!\n")
            flagEst = True
        #else:
            #print("Error de ruta 2, Auto Estancado")
        return [flagEner, flagN3, flagEst, node2, node3, eA, time, vec]
    
    #Funcion para tomar decision en funcion de la carga restante
    def mercaRest(self, nodoSig):
        if self.merca >= nodoSig.demanda:
            flagCarga = True
        else:
            flagCarga = False
        return flagCarga
    
    
    def factibleNode(self, node2):
    
        #velocidad del carro
        v = float(self.world.velocity)
        recRate = float(self.world.recRate)
        conRate = float(self.world.conRate)
        cargaE = float(self.world.cargaE)
        sano = None
        #Coordenadas
        if type(self.node()) == list:
            coord1 = self.node()[0].getCoord()
        else:
            coord1 = self.node().getCoord()
        #print(node2)
        if type(node2) == list:
            coord2 = node2[0].getCoord()
        else:
            coord2 = node2.getCoord()
        time = cop.deepcopy(self.time)
        energia = cop.deepcopy(self.energia)
        flagGeneral = False
        flagDep = False
        flagEst = False
        flagN3 = False
        #revisa que el nodo2 seleccionado no este entre los visitados
        if node2 != None:
            if type(node2) == list:
                flagCarga = self.mercaRest(node2[0])
            else:
                flagCarga = self.mercaRest(node2)
        else:
            flagCarga = False
            
        #Primera Decicion: Carga de Mercancia
        if flagCarga:
            #print("\nEntra aqui en la factibilidad de carga:\n")
            #FUNCION OBJETIVO:
            ####################################################################################################
            dist1 = (self.world.costoVeh) + (self.world.costoDis * euclidean(coord1, coord2))
            ####################################################################################################   
            estNodo1 = self.lookCharge(coord1)
            estNodo2 = self.lookCharge(coord2)
            energiaR = self.energiaRst(node2)
            energia = energiaR[5]
            time += energiaR[6]
            nodo3 = energiaR[4]
            nodesCerca = energiaR[7]
            tiempoR = self.evalTiempo(node2, dist1, nodo3, nodesCerca, time)
            time += tiempoR[4]
            #print("\nEnergia: ", energiaR[0], " ", energiaR[1], " ", energiaR[2], "\n")
            #print("energia Actual: ", self.energia, " ", dist1*self.world.conRate)
            #print("\nTiempo: ", tiempoR[0], " ", tiempoR[1], " ", tiempoR[2], " ", tiempoR[3], "\n")
            ####################################################################################################
            #Segunda Decicion B: Ventanas de Tiempo
            #Cuando esta dentro de la ventana de tiempo
            if energiaR[0] and energiaR[1] and not energiaR[2]:
                #Deciciones sobre ventanas de tiempo
                #viajar nodo2 y viajar nodo3
                flagEst = False
                if tiempoR[0] and tiempoR[1] and not tiempoR[2] and not tiempoR[3]:
                    flagGeneral = True
                    flagN3 = True
                    flagDep = False
                elif tiempoR[0] and not tiempoR[1] and not tiempoR[2] and not tiempoR[3]:
                    flagGeneral = True
                    flagN3 = False
                    flagDep = False
                elif not tiempoR[0] and not tiempoR[1] and tiempoR[2] and not tiempoR[3]:
                    flagGeneral = False
                    flagN3 = False
                    flagDep = False
                elif not tiempoR[0] and not tiempoR[1] and tiempoR[2] and tiempoR[3]:
                    flagGeneral = False
                    flagN3 = False
                    flagDep = True
                    
            elif energiaR[0] and not energiaR[1] and not energiaR[2]:
                #Decisiones sobre ventanas de tiempo
                #viajar nodo2
                flagN3 = False
                flagEst = False
                if tiempoR[0] and not tiempoR[2] and not tiempoR[3]:
                    flagGeneral = True
                    flagDep = False
                elif not tiempoR[0] and tiempoR[2] and not tiempoR[3]:
                    flagGeneral = False
                    flagDep = False
                elif not tiempoR[0] and tiempoR[2] and tiempoR[3]:
                    flagGeneral = False
                    flagDep = True
                    
            elif not energiaR[0] and not energiaR[1] and energiaR[2]:
                #print("\nEnergias para ir a una estacion")
                #print("\nViaje a una estacion cercana:\n")
                #print("sano marico")
                sano = True
                flagGeneral = False
                flagN3 = False
                flagEst = True
                if tiempoR[0] and tiempoR[1] and not tiempoR[2] and not tiempoR[3]:
                    flagDep = False
                elif tiempoR[0] and not tiempoR[1] and not tiempoR[2] and not tiempoR[3]:
                    flagDep = False
                elif not tiempoR[0] and not tiempoR[1] and tiempoR[2] and not tiempoR[3]:
                    flagDep = False
                elif not tiempoR[0] and not tiempoR[1] and tiempoR[2] and tiempoR[3]:
                    flagDep = True
            else:
                flagGeneral = False
                flagN3 = False
                flagEst = False
                flagDep = True
        #Else de la Primera Decicion
        else:
            #print("\n Deposito: asxasxas\n")
            energiaR = self.energiaRst(node2)
            estNodo1 = self.lookCharge(coord1)
            nodo3 = energiaR[4]
            energia = energiaR[5]
            time += energiaR[6]
            dist1 = (self.world.costoVeh) + (self.world.costoDis * euclidean(coord1, coord2))
            nodesCerca = energiaR[7]
            tiempoR = self.evalTiempo(node2, dist1, nodo3, nodesCerca, time)
            time += tiempoR[4]
            flagGeneral = False
            flagN3 = False
            flagDep = True
            if energiaR[2]:
                flagEst = True
            else:
                flagEst = False
            #print("\nFlagGeneral: ", flagGeneral)
            #print("\nFlagN3: ", flagN3)
            #print("\nFlagDep: ", flagDep)
            #print("\nFlagEst: ", flagEst)
        if sano != None and sano == True:
            #print([flagGeneral, flagN3, flagEst, flagDep, time, energia, estNodo1, nodo3])
            sano = False
        return [flagGeneral, flagN3, flagEst, flagDep, time, energia, estNodo1, nodo3]

    def can_move(self):
        #Aqui van las condiciones donde se ven los 1/4nodes mas cercanos y se evaluan con factibleNode para ver que
        #la hormiga se pueda mover o no
        flagEst = False
        flagMerca = False
        estC = None
        notF = []
        notF.clear()
        selec = []
        selec.clear()
        selec2 = []
        selec2.clear()
        if self.merca >= min(self.unvisited, key = lambda x : x.demanda).demanda:
            flagMerca = True
            selec = self.Selec14(self.node())
            selec2 = cop.deepcopy(selec)
            #print("\nselec: ", len(selec), "\n")
            if len(selec) > 0:
                sel = []
                sel.clear()
                
                for i in range(len(selec)):
                    nod = selec[i]
                    if type(nod) == list:
                        fl = nod[0].flagV
                    else:
                        fl = nod.flagV
                    #print("\nEvaluacion nodo: ", i, "\n")
                    fact1 = self.factibleNode(selec[i])
                    if fact1[0] and not fl:
                        #print("\nfactible:\n")
                        sel.append(selec[i])
                        #print("\nflagGeneral: ", self.factibleNode(selec[i])[0])
                        #print("\nflagN3: ", self.factibleNode(selec[i])[1])
                        #print("\nflagEst: ", self.factibleNode(selec[i])[2])
                        #print("\nflagDep: ", self.factibleNode(selec[i])[3])
                    elif not fact1[0] and fact1[2] and not fact1[3]:
                        notF.append(selec[i])
                #print("\nLength de notF:", len(notF))
                #print("\nseleccion: ", len(sel))
                if self.node().tipo == 'e' or self.node() == self.start:
                    flagEst = True
                    if len(self.visited) > 1:
                        ultNodo = self.visited[-2]
                        #print("\nNodo anterior a mi: ", ultNodo)
                        #print("\nNodo actual: ", self.node())
                        estC = self.lookCharge(self.node().getCoord())
                        while estC == ultNodo or estC == self.node():
                            estC = self.lookCharge(self.node().getCoord())
                else:
                    flagEst = False
                    if len(self.visited) > 1:
                        ultNodo = self.visited[-2]
                        #print("\nNodo anterior a mi: ", ultNodo)
                        #print("\nNodo actual: ", self.node())
                        estC = self.lookCharge(self.node().getCoord())
                        while estC == ultNodo or estC == self.node():
                            estC = self.lookCharge(self.node().getCoord())
                return [sel, notF, flagEst, flagMerca, estC, selec2]
            else:
                return [[], notF, flagEst, flagMerca, estC, selec2]
        else:
            flagMerca = False
            #print("\nOutOfMerca...")
            return [[], notF, flagEst, flagMerca, estC, selec2]
        
    def choose_move(self, choices):
        if len(choices) == 0:
            return None
        if len(choices) == 1:
            if type(choices[0]) == list:
                return choices[0][0]
            else:
                return choices[0]
        
        # Find the weight of the edges that take us to each of the choices.
        weights = []
        for move in choices:
            if type(self.node()) == list:
                #print("\nEntro por aqui, tipo nodo = list")
                if type(move) == list:
                    #print("\nMove Lista:")
                    #print("\n", move[0].tipo)
                    edge = self.world.edges[self.node()[0], move[0]]
                else:
                    #print("\nMove not lista")
                    #print("\n", move)
                    edge = self.world.edges[self.node()[0], move]
            else:
                #print("\nEntro por aqui, tipo nodo not list")
                if type(move) == list:
                    #print("\nMove Lista:")
                    #print("\n", move[0].tipo)
                    edge = self.world.edges[self.node(), move[0]]
                else:    
                    #print("\nMove not lista")
                    #print("\n", move)
                    edge = self.world.edges[self.node(), move]
            weights.append(self.weigh(edge))
        
        # Choose one of them using a weighted probability.
        total = sum(weights)
        cumdist = list(itertools.accumulate(weights)) + [total]
        choice = choices[bisect.bisect(cumdist, random.random() * total)]
        #print("\nescogio: ", choice, "\n")
        return choice

    def move(self):
        remaining = self.can_move()
        #print("\nLongitud de remaining: ", len(remaining))
        #if len(remaining) == 0:
            #print("\nNo existen nodos factibles: \n")
            #print("\nMerca: ", self.merca)
        choice = self.choose_move(remaining[0])
        #print("\nescogimiento: ", choice)
        if choice != None:
            fact = self.factibleNode(choice)
            #print(fact)
        elif len(remaining[1]) > 0 and remaining[3]:
            #print("\nEleccion de nodos infactibles, Continua a las tomas de desicion:\n")
            choice = self.choose_move(remaining[1])
            #print("\nescogimiento: ", choice)
            fact = self.factibleNode(choice)
        else:
            #print("\nchoice es una lista vacia:\n")
            edgeDep = self.insert_depot()
            #print("\nedgeDep: ", edgeDep)
            #if edgeDep != None:
            #    if type(edgeDep) == Ant:
                    #print("Hormiga #1")
            return self
        #print("\nTomas de decision: \n")
        if fact[0] and fact[1] and not fact[2] and not fact[3]:
            #viaje a nodo2 y nodo3 sin recarga
            #print("\nviaje a nodo2 y nodo3 sin recarga:\n")
            self.energia = fact[5]
            edgeN2 = self.make_move(choice)
            if type(choice) == list:
                self.merca -= choice[0].demanda
            else:
                self.merca -= choice.demanda
            edgeN3 = self.make_move(fact[7])
            if type(fact[7]) == list:
                self.merca -= fact[7][0].demanda
            else:
                self.merca -= fact[7].demanda
            #print("\nedgeN2: ", edgeN2)
            #print("\nedgeN3: ", edgeN3)
            #if edgeN2 != None:
                #if type(edgeN2) == Ant:
                    #print("Hormiga #2.1")
                #if edgeN3 != None:
                #    if type(edgeN3) == Ant:
                        #print("Hormiga #2.2")
            return self
        
        elif fact[0] and not fact[1] and not fact[2] and not fact[3]:
            #viaje a nodo2 sin recarga
            #print("Decision nro. 1: ")
            #print("\nviaje a nodo2 sin recarga:\n")
            self.energia = fact[5]
            edgeN2 = self.make_move(choice)
            if type(choice) == list:
                self.merca -= choice[0].demanda
            else:
                self.merca -= choice.demanda
            #print("\nedgeN2: ", edgeN2)
            #if edgeN2 != None:
            #    if type(edgeN2) == Ant:
                    #print("Hormiga #3")
                    ##print(choice.tipo)
                    #print("##########")
            return self

        elif not fact[0] and not fact[1] and fact[2] and not fact[3]:
            #print("Decision nro. 2: ")
            #print("\nViaje a una estacion cercana 2: ")
            self.energia = fact[5]
            edgeEst = self.make_move(fact[6])
            #print("\nedgeEst: ", edgeEst)
            #if edgeEst != None:
            #    if type(edgeEst) == Ant:
                    #print("Hormiga #4")
            return self
        elif not fact[0] and fact[2] and fact[3]:
            #print("Decision nro. 3: ")
            #print("\nviaje a deposito con recarga\n")
            self.energia = fact[5]
            edgeEst = self.make_move(fact[6])
            #print("\nedgeEst: ", edgeEst)
            #if edgeEst != None:
            #    if type(edgeEst) == Ant:
                    #print("Hormiga #5")
            return self
        elif not fact[0] and not fact[2] and fact[3]:
            #print("Decision nro. 4: ")
            #print("\nviaje a deposito sin recarga\n")
            self.energia = fact[5]
            edgeDep = self.insert_depot()
            #print("\nedgeDep: ", edgeDep)
            #if edgeDep != None:
            #    if type(edgeDep) == Ant:
                    #print("Hormiga #6")
            return self
        else:
            #print("\nError de factibilidad en move(): \n")
            raise KeyboardInterrupt
        self.time = fact[4]
        #print("SANO ###############################################")
    
    def lookCharge(self, coord):
        veDist = []
        for i in range(len(self.world.estaciones)):
            if coord != self.world.estaciones[i].getCoord():
                veDist.append(euclidean(coord, self.world.estaciones[i].getCoord()))
                #print(str(veDist[-1]), "\t\t", str(estaciones[i]), "\n")
        menor = min(veDist)
        minInd = veDist.index(menor)
        return [self.world.estaciones[minInd], menor]
    
    def look_charge(self, coord):
        veDist = []
        #print(self.node())
        for i in range(len(estaciones[inst])):
            if estaciones[inst][str(i)] != coord:
                veDist.append(euclidean(coord, estaciones[inst][str(i)]))
                #print(str(veDist[-1]), "\t\t", str(estaciones[i]), "\n")
        menor = min(veDist)
        minInd = veDist.index(menor)
        return [minInd, estaciones[inst][str(minInd)], menor]
        
    def insert_depot(self):
        #print("\nInserta depot: \n")
        node1 = self.node()
        if type(node1) == list:
            node1 = node1[0]
        time = 0
        coordDep = self.start.getCoord()
        coord1 = node1.getCoord()
        dist1 = (self.world.costoVeh) + (self.world.costoDis * euclidean(coord1, coordDep))
        enerDep = self.world.conRate * dist1
        if node1.tipo == 'e':
            #print("\nse encuentra en una estacion: ")
            if self.energia <= enerDep and enerDep <= self.world.cargaE:
                dif = enerDep - self.energia
                time += dif/self.world.recRate
                self.energia = enerDep
                self.make_move(None)
            elif self.energia <= enerDep and enerDep > self.world.cargaE:
                Est = self.look_charge(coord1)
                nodEst = self.world.estaciones[Est[0]]
                #print("\n estacion mas cercana: ", Est[0])
                coordEst = nodEst.getCoord()
                distNE = (self.world.costoVeh) + (self.world.costoDis * euclidean(coord1, coordEst))
                #print("\ndistancia: ", distNE)
                enerEst = distNE * self.world.conRate
                #print("\nEnergia actual: ", self.energia)
                #print("\nEnergia necesaria: ", enerEst)
                if self.energia <= enerEst and enerEst <= self.world.cargaE:
                    #print("\nBuscando otra estacion:\n")
                    dif = enerEst - self.energia
                    time += dif/self.world.recRate
                    self.energia = enerEst
                    self.make_move(nodEst)
                    self.insert_depot()
                elif self.energia >= enerEst:
                    #print("\nViaja a una estacion cercana: ")
                    self.make_move(nodEst)
                    self.insert_depot()
                #else:
                    #print("\nError, No puede ir a ninguna estacion: 1\n")
            elif self.energia >= enerDep:
                self.make_move(None)
        elif node1.tipo == 'c':
            if self.energia <= enerDep:
                Est = self.look_charge(coord1)
                nodEst = self.world.estaciones[Est[0]]
                #print("\Se encuentra en un cliente: ")
                #print("\n estacion mas cercana: ", Est[0])
                coordEst = nodEst.getCoord()
                distNE = (self.world.costoVeh) + (self.world.costoDis * euclidean(coord1, coordEst))
                #print("\ndistancia: ", distNE)
                enerEst = distNE * self.world.conRate
                #print("\nEnergia actual: ", self.energia)
                #print("\nEnergia necesaria: ", enerEst)
                if self.energia <= enerEst and enerEst <= self.world.cargaE:
                    #print("\nBuscando otra estacion:\n")
                    dif = enerEst - self.energia
                    time += dif/self.world.recRate
                    self.energia = enerEst
                    self.make_move(nodEst)
                    self.insert_depot()
                elif self.energia >= enerEst:
                    #print("\nViaja a una estacion cercana: ")
                    self.make_move(nodEst)
                    self.insert_depot()
                #else:
                    #print("\nError, No puede ir a ninguna estacion: 2\n")
            elif self.energia >= enerDep:
                self.make_move(None)
        return self

    def remaining_moves(self):
        return self.unvisited

    def make_move(self, dest):
        
        # Since self.node simply refers to self.visited[-1], which will be
        # changed before we return to calling code, store a reference now.
        ori = self.node()

        # When dest is None, all nodes have been visited but we may not
        # have returned to the node on which we started. If we have, then
        # just do nothing and return None. Otherwise, set the dest to the
        # node on which we started and don't try to move it from unvisited
        # to visited because it was the first one to be moved.
        if dest is None:
            #print("\nRegreso al depot:")
            dest = self.start   # last move is back to the start
            #print(dest)
            self.visited.append(dest)
            self.time = 0
            self.energia = cop.copy(self.world.cargaE)
            self.merca = cop.copy(self.mercaMax)
            if type(ori) == list:
                if type(dest) == list:
                    if ori[0] != dest[0]:
                        edge = self.world.edges[ori[0], dest[0]]
                        #print("\n edge = None: 1a\n")
                        #print(self.world.edges)
                    else:
                        #print("\n edge = None: 1b\n")
                        #print(self.world.edges)
                        edge = None
                else:
                    if ori[0] != dest:
                        edge = self.world.edges[ori[0], dest]
                        #print("\n edge = None: 2a\n")
                        #print(self.world.edges)
                    else:
                        #print("\n edge = None: 2b\n")
                        #print(self.world.edges)
                        edge = None
            else:
                if type(dest) == list:
                    if ori != dest[0]:
                        edge = self.world.edges[ori, dest[0]]
                        #print("\n edge = None: 3a\n")
                        #print(self.world.edges)
                    else:
                        #print("\n edge = None: 3b\n")
                        #print(self.world.edges)
                        edge = None
                else:
                    if ori != dest:
                        edge = self.world.edges[ori, dest]
                        #print("\n edge = None: 4a\n")
                        #print(edge)
                        #print(self.world.edges)
                    else:
                        #print("\n edge = None: 4b\n")
                        #print(self.world.edges)
                        edge = None
                        #print("Regreso al depot de makeMove: ", edge, " ", self)
            if edge != None and type(edge) == Edge and type(edge) != Ant:
                self.traveled.append(edge)
            return self
        else:
            if type(dest) == list:
                #print("\nDestino type lista :", dest)
                #print("\ntipo de nodo destino :", dest[0].tipo) 
                self.visited.append(dest[0])
                dest[0].flagV = True
            else:
                #print("\nDestino type nodo:", dest)
                #print("\ntipo de nodo destino :", dest.tipo)
                self.visited.append(dest)
                dest.flagV = True
        if type(ori) == list:
            coord1 = ori[0].getCoord()
            if type(dest) == list:
                coord2 = dest[0].getCoord()
                #print("1", ori[0]==dest[0])
                if ori[0]==dest[0]:
                    cosaquenohacenada = None
                    #print("ori: ", ori[0].getCoord())
                    #print("dest: ", dest[0].getCoord())
                else:    
                    edge = self.world.edges[ori[0], dest[0]]
                    if type(edge) == Edge and type(edge) != Ant:
                        self.traveled.append(edge)
                    dist = (self.world.costoVeh) + (self.world.costoDis * euclidean(coord1, coord2))
                    self.energia -= self.world.conRate * dist
                    #print("dist: ", dist)
                    self.distance += dist
                    #print("distance: ", self.distance)
                    self.cont += 1
                    #print(self.cont)
                    return self
            else:
                coord2 = dest.getCoord()
                #print("2", ori[0]==dest)
                if ori[0]==dest:
                    cosaquenohacenada = None
                    #print("ori: ", ori[0].getCoord())
                    #print("dest: ", dest.getCoord())
                else:    
                    edge = self.world.edges[ori[0], dest]
                    if type(edge) == Edge and type(edge) != Ant:
                        self.traveled.append(edge)
                    dist = (self.world.costoVeh) + (self.world.costoDis * euclidean(coord1, coord2))
                    self.energia -= self.world.conRate * dist
                    #print("dist: ", dist)
                    self.distance += dist
                    #print("distance: ", self.distance)
                    self.cont += 1
                    #print(self.cont)
                    return self
        else:
            coord1 = ori.getCoord()
            if type(dest) == list:
                coord2 = dest[0].getCoord()
                #print("3", ori==dest[0])
                if ori == dest[0]:
                    cosaquenohacenada = None
                    #print("ori: ", ori.getCoord())
                    #print("dest: ", dest[0].getCoord())
                else:    
                    edge = self.world.edges[ori, dest[0]]
                    if type(edge) == Edge and type(edge) != Ant:
                        #print(type(edge))
                        self.traveled.append(edge)
                        #print("Sano: ", self.traveled[-1])
                    dist = (self.world.costoVeh) + (self.world.costoDis * euclidean(coord1, coord2))
                    self.energia -= self.world.conRate * dist
                    #print("dist: ", dist)
                    self.distance += dist
                    #print("distance: ", self.distance)
                    self.cont += 1
                    #print(self.cont)
                    return self
            else:
                coord2 = dest.getCoord()
                #print("4", ori==dest)
                if ori == dest:
                    cosaquenohacenada = None
                    #print("ori: ", ori.getCoord())
                    #print("dest: ", dest.getCoord())
                else:    
                    edge = self.world.edges[ori, dest]
                    if type(edge) == Edge and type(edge) != Ant:
                        self.traveled.append(edge)
                    dist = (self.world.costoVeh) + (self.world.costoDis * euclidean(coord1, coord2))
                    self.energia -= self.world.conRate * dist
                    #print("dist: ", dist)
                    self.distance += dist
                    #print("distance: ", self.distance)
                    self.cont += 1
                    #print(self.cont)
                    return self

    def weigh(self, edge):
        
        pre = 1 / (edge.length or 1)
        post = edge.pheromone
        return post ** self.alpha * pre ** self.beta
    
    def areUnv(self):
        for i in self.unvisited:
            if not i.flagV:
                return True
```

```python
#Solver
class Solver:
    
    def __init__(self, alpha, beta, merca, energia, costoVeh, costoDis, inst, **kwargs):
        self.alpha = alpha
        self.beta = beta
        self.rho = kwargs.get('rho', 0.8)
        self.q = kwargs.get('Q', 1)
        self.t0 = kwargs.get('t0', .01)
        self.limit = kwargs.get('limit', 100)
        self.ant_count = kwargs.get('ant_count', 10)
        self.elite = kwargs.get('elite', .5)
        self.merca = merca
        self.energia = energia
        self.costoVeh = costoVeh
        self.costoDis = costoDis
        
    def create_colony(self, world):
        
        if self.ant_count < 1:
            return self.round_robin_ants(world, len(world.nodes))
        return self.random_ants(world, self.ant_count)
        
    def reset_colony(self, colony):
        
        for ant in colony:
            #Cambio en la siguiente linea, starting node world.nodes[0]
            ant.initialize(ant.world, start=ant.world.nodes[0])
        
    def aco(self, colony):
        
        cont = self.find_solutions(colony)
        self.global_update(colony)
        return sorted(colony)[cont]
        
    def solve(self, world):
        
        world.reset_pheromone(self.t0)
        global_best = None
        colony = self.create_colony(world)
        for i in range(self.limit):
            self.reset_colony(colony)
            local_best = self.aco(colony)
            if global_best is None or local_best.distance < global_best.distance:
                global_best = cop.copy(local_best)
            self.trace_elite(global_best)
        #global_best.append(world.data(0))
        return global_best
    
    def solutions(self, world):
        
        world.reset_pheromone(self.t0)
        global_best = None
        colony = self.create_colony(world)
        for i in range(self.limit):
            self.reset_colony(colony)
            local_best = self.aco(colony)
            if global_best is None or local_best < global_best:
                global_best = cop.copy(local_best)
                yield global_best
            self.trace_elite(global_best)
    
    def round_robin_ants(self, world, count):
        
        starts = world.nodes
        n = len(starts)
        return [
            Ant(self.alpha, self.beta, self.merca, self.energia, self.costoVeh, self.costoDis).initialize(
                world, start=starts[i % n])
            for i in range(count)
        ]
        
    def random_ants(self, world, count, even=False):
        
        ants = []
        starts = world.nodes
        n = len(starts)
        if even:
            # Since the caller wants an even distribution, use a round-robin 
            # method until the number of ants left to create is less than the
            # number of nodes.
            if count > n:
                for i in range(self.ant_count // n):
                    ants.extend([
                        Ant(self.alpha, self.beta, self.merca, self.energia, self.costoVeh, self.costoDis, inst).initialize(
                            world, start=starts[j])
                        for j in range(n)
                    ])
            # Now (without choosing the same node twice) choose the reamining
            # starts randomly.
            ants.extend([
                Ant(self.alpha, self.beta, self.merca, self.energia, self.costoVeh, self.costoDis, inst).initialize(
                    world, start=starts.pop(random.randrange(n - i)))
                for i in range(count % n)
            ])
        else:
            # Just pick random nodes.
            ants.extend([
                Ant(self.alpha, self.beta, self.merca, self.energia, self.costoVeh, self.costoDis, inst).initialize(
                    world, start=starts[random.randrange(n)]) 
                for i in range(count)
            ])
        return ants

    def find_solutions(self, ants):
        
        # This loop occurs exactly as many times as there are ants times nodes,
        # but that is only because every ant must visit every node. It may be
        # more efficient to convert it to a counting loop... but what 
        # flexibility would we loose in terms of extending the solver class?
        ants_done = 0
        while ants_done < len(ants):
            ants_done = 0
            contAnt = 0
            for ant in ants:
                for node in ant.world.nodes:
                    node.flagV = False
                    #print(node.flagV)
                #print("ENTRA1: ")
                cont1 = 0
                cont2 = 0
                notf1 = 0
                while ant.areUnv:
                    can = ant.can_move()
                    if len(can[0]) > 0 or len(can[1]) > 0:
                        #print("#############################")
                        #print("ENTRA2: ")
                        #print(len(can[0]), " ", len(can[1]), " ", len(can[-1]))
                        #print("#############################")
                        ant = ant.move()
                        edge = ant.traveled
                        for i in ant.traveled:
                            self.local_update(i)
                        if len(can[1]) == notf1:
                            cont1 += 1
                        else:
                            cont1 = 0
                            
                        if cont1 == 20:
                            #print("\nNo visito todos...\n")
                            contAnt += 1
                            break
                    else:
                        #print("#############################")
                        #print("\nENTRA3: ")
                        #print(len(can[0]), " ", len(can[1]), " ", len(can[-1]))
                        #print("#############################")
                        ant = ant.move()
                        if len(can[-1]) == 0:
                            cont2 += 1
                        else:
                            cont2 = 0
                        
                        if cont2 == 20:
                            break
                    notf1 = len(can[1])

                #print("SALIO:")
                ants_done += 1
                print(ants_done)
        return contAnt

    def local_update(self, edge):
        #print("\ntype edge: ", type(edge))
        #print("\nedge: ", edge)
        if type(edge) != None and edge != None:
            #print("\no es None\n")
            if type(edge) == tuple:
                #print("\nes tupla\n")
                edge[0].pheromone = max(self.t0, edge[0].pheromone * self.rho)
            else:
                #print("\nes edge\n")
                edge.pheromone = max(self.t0, edge.pheromone * self.rho)

    def global_update(self, ants):
        ants = sorted(ants)[:len(ants) // 2]
        print("\nLongitud vector hormigas:", len(ants))
        for a in ants:
            if a.distance > 0:
                p = self.q / a.distance
                for edge in a.path():
                    if edge != None:
                        if type(edge) == tuple:
                            edge[0].pheromone = max(self.t0, (1 - self.rho) * edge[0].pheromone + p)
                        else:    
                            edge.pheromone = max(self.t0, (1 - self.rho) * edge.pheromone + p)

    def trace_elite(self, ant):
        if self.elite:
            if ant.distance > 0:
                p = self.elite * self.q / ant.distance
                for edge in ant.path():
                    if edge != None:
                        if type(edge) == tuple:
                            edge[0].pheromone += p
                        else:    
                            edge.pheromone += p
```
